# Task: City Analysis

## Identify the city with the highest number of restaurants in the dataset.

## Calculate the average rating for restaurants in each city.

## Determine the city with the highest average rating.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset .csv')
```

```python
df.head()
```

```python
df.info()
```

## Identify the city with the highest number of restaurants in the dataset.

```python
# Ensure column names are stripped of any extra spaces
df.columns = df.columns.str.strip()

# Use value_counts to find the most frequent city
most_frequent_city = df['City'].value_counts().index[0]
print(f'The city with the highest number of restaurants is {most_frequent_city}.')
```

## Calculate the average rating for restaurants in each city.

```python
# Group by 'City' and calculate the average of 'Aggregate rating'
average_rating_by_city = df.groupby('City')['Aggregate rating'].mean()

# Create a DataFrame from the Series and sort it by average rating in descending order
average_rating_df = average_rating_by_city.reset_index()
average_rating_df.columns = ['City', 'Average_rating']
average_rating_df = average_rating_df.sort_values(by='Average_rating', ascending=False)

print(average_rating_df)
```

```python
# Group by 'City' and calculate the average of 'Aggregate rating'
average_restaurant_rating = df.groupby('City')['Aggregate rating'].mean().reset_index()
average_restaurant_rating.columns = ['City', 'Average Rating']

# Sort values by 'Average Rating'
average_restaurant_rating = average_restaurant_rating.sort_values(by='Average Rating', ascending=False)

# Plotting
plt.figure(figsize=(14, 8))
ax = sns.barplot(x='City', y='Average Rating', data=average_restaurant_rating, palette='viridis')

# Add labels on bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

# Adding title and labels
plt.title('Average Rating for Restaurants in Each City', fontsize=16)
plt.xlabel('City', fontsize=14)
plt.ylabel('Average Rating', fontsize=14)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

plt.tight_layout()
plt.show()
```

## Determine the city with the highest average rating

```python
# Group by 'City' and calculate the average of 'Aggregate rating'
average_restaurant_rating = df.groupby('City')['Aggregate rating'].mean()

# Find the city with the highest average rating
highest_rated_city = average_restaurant_rating.nlargest(1)

# Extract the city name and the highest average rating
city_name = highest_rated_city.index[0]
rating_value = highest_rated_city.values[0]

print(f'The city with the highest average rating is {city_name} with an average rating of {rating_value:.2f}.')
```

## Task: Price Range Distribution

## Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.

## Calculate the percentage of restaurants in each price range category.

```python
#import the library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data set 

df = pd.read_csv('Dataset .csv')
```

```python
df.head()
```

```python
df.info()
```

## Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.

```python
# Ensure column names are stripped of any extra spaces
df.columns = df.columns.str.strip()

# Count the frequency of each 'Price range'
price_range_counts = df['Price range'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
price_range_counts.plot(kind='bar', color='DarkRed', edgecolor='Black')

# Adding labels and title
plt.title('Price Range Distribution')
plt.xlabel('Price Range')
plt.ylabel('Frequency')

# Adding value labels on the bars
for index, value in enumerate(price_range_counts):
    plt.text(index, value, str(value), ha='center', va='bottom')

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.tight_layout()
plt.show()
```

## Calculate the percentage of restaurants in each price range category.

```python
# Group by 'Price range' and count occurrences
price_range_counts = df.groupby('Price range').size().reset_index(name='Count')

print(price_range_counts)
```

```python
# Calculate the count of restaurants in each price range
price_range_counts = df.groupby('Price range').size().reset_index(name='Count')

# Calculate the total number of restaurants
total_restaurants = price_range_counts['Count'].sum()

# Calculate the percentage of restaurants in each price range category
price_range_counts['Percentage'] = (price_range_counts['Count'] / total_restaurants) * 100

# Print the result
print("Percentage of restaurants in each price range category:")
for index, row in price_range_counts.iterrows():
    print(f"{row['Price range']}: {row['Percentage']:.2f}%")
```

```python
# Calculate the count of restaurants in each price range
price_range_counts = df['Price range'].value_counts().reset_index(name='Count')

# Calculate the total number of restaurants
total_restaurants = price_range_counts['Count'].sum()

# Calculate the percentage of restaurants in each price range category
price_range_counts['Percentage'] = (price_range_counts['Count'] / total_restaurants) * 100

# Define a list of four named colors for the pie chart
colors = ['lightcoral', 'lightskyblue', 'lightgreen', 'lightpink']

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
ax.pie(price_range_counts['Percentage'], labels=price_range_counts['Price range'], autopct='%1.1f%%', colors=colors)

# Adding title
plt.title('Percentage of Restaurants in Each Price Range')

plt.show()
```
